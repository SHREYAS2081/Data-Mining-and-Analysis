#### Making a new dataframe for the total number of Votes per City:

```python
f=c.groupby(['City']).sum()
s=pd.DataFrame(f['Votes'])
s
```

____
#### To make a new dataframe showing  mean Aggregate rating per City:

```python
l=c.groupby(['City']).mean()
m=pd.DataFrame(l['Aggregate rating'])
m.reset_index()
```

____
#### Plotting graph for the cities and there mean Aggregate rating:
